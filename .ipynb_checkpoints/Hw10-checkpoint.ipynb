```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.applications import VGG16
from keras.layers import Dense, Flatten, Dropout, Input, Conv2D, MaxPooling2D
from keras import optimizers
from keras import losses
from keras import metrics
from keras.utils import to_categorical
from keras.datasets import fashion_mnist
```

Завантаження даних

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Розмірність даних (shape)
print("Shape of x_train:", x_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of x_test:", x_test.shape)
print("Shape of y_test:", y_test.shape)

# Кількость зразків у наборі даних (len)
print("Number of samples in x_train:", len(x_train))
print("Number of samples in y_train:", len(y_train))
print("Number of samples in x_test:", len(x_test))
print("Number of samples in y_test:", len(y_test))
```

```python
# Заголовок для всього набору зображень
plt.figure(figsize=(7, 5))
plt.suptitle("Перші 15 зображень з fashion MNIST", fontsize=9)

# Відображення перших 15 зображень
for i in range(15):
    plt.subplot(3, 5, i + 1)  # 3 рядки, 5 стовпців, індекс поточного графіку
    plt.imshow(x_train[i], cmap=plt.get_cmap('gray'))
    plt.axis('off')  # Вимкнути вісі для зручності
    plt.title(f"Зображення {i+1}", fontsize=9)

plt.tight_layout()
plt.show()
```

```python
model = Sequential([
    Input(shape=(28, 28, 1)),  # Вхідний шар
    Conv2D(32, kernel_size=(3, 3), activation='relu'),  # Перший згортковий шар
    MaxPooling2D(pool_size=(2, 2)),  # Перший шар пулінгу
    Conv2D(64, kernel_size=(3, 3), activation='relu'),  # Другий згортковий шар
    MaxPooling2D(pool_size=(2, 2)),  # Другий шар пулінгу
    Conv2D(64, kernel_size=(3, 3), activation='relu'),  # Третій згортковий шар
    Flatten(),  # Розгортання для передачі до повністю з'єднаного шару
    Dense(128, activation='relu'),  # Повністю з'єднаний шар
    Dropout(0.2),  # Застосування випадкового вимикання для уникнення перенавчання
    Dense(10, activation='sigmoid')  # Вихідний шар з функцією активації softmax для класифікації
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()  # Вивести огляд архітектури моделі
```

```python
history = model.fit(x_train.reshape(-1, 28, 28, 1), y_train, epochs=10, batch_size=64, validation_data=(x_test.reshape(-1, 28, 28, 1), y_test))
```

```python
# Графік точності навчання та валідації
plt.plot(history.history['accuracy'], label='Точність навчання')
plt.plot(history.history['val_accuracy'], label = 'Точність валідації')
plt.xlabel('Епохи')
plt.ylabel('Точність')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.title('Графік точності навчання та валідації')
plt.show()

# Графік втрат навчання та валідації
plt.plot(history.history['loss'], label='Втрати навчання')
plt.plot(history.history['val_loss'], label = 'Втрати валідації')
plt.xlabel('Епохи')
plt.ylabel('Втрати')
plt.ylim([0, 2])
plt.legend(loc='upper right')
plt.title('Графік втрат навчання та валідації')
plt.show()
```

```python
test_loss, test_accuracy = model.evaluate(x_test.reshape(-1, 28, 28, 1), y_test, verbose=2)

print(f"Втрати на тестовому наборі: {test_loss}")
print(f"Точність на тестовому наборі: {test_accuracy}")
```

```python
# Завантаження зважень моделі VGG16 без верхніх повністю з'єднаних шарів
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

x_train_resized = np.pad(x_train, ((0,0),(2,2),(2,2)), mode='constant')
x_test_resized = np.pad(x_test, ((0,0),(2,2),(2,2)), mode='constant')

# Додавання каналів кольору
x_train_rgb = np.expand_dims(x_train_resized, axis=-1)
x_train_rgb = np.repeat(x_train_rgb, 3, axis=-1)

x_test_rgb = np.expand_dims(x_test_resized, axis=-1)
x_test_rgb = np.repeat(x_test_rgb, 3, axis=-1)

# Замороження ваг моделі VGG16
base_model.trainable = False

# Додавання повністю з'єднаного шару та шару вихідного класифікатора
model = Sequential([
    base_model,
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()  # Вивести огляд архітектури моделі
```

```python
!pip install seaborn

from sklearn.metrics import confusion_matrix
import seaborn as sns

# Передбачення класів для тестових даних
y_pred = model.predict_classes(x_test_rgb)

# Розрахунок матриці плутанини
conf_matrix = confusion_matrix(y_test, y_pred)
# Візуалізація матриці плутанини
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Blues', xticklabels=range(10), yticklabels=range(10))
plt.xlabel('Предсказаний клас')
plt.ylabel('Справжній клас')
plt.title('Матриця плутанини')
plt.show()

# Випадкові приклади зображень з передбаченими та справжніми мітками
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.imshow(x_test_rgb[i], cmap='gray')
    plt.title(f"П: {y_pred[i]}, С: {y_test[i]}")
    plt.axis('off')
plt.suptitle('Приклади зображень з передбаченими та справжніми мітками')
plt.show()
```

```python
# Оцінка втрат та точності на тестовому наборі даних
test_loss, test_accuracy = model.evaluate(x_test_rgb, y_test, verbose=2)

print(f"Втрати на тестовому наборі: {test_loss}")
print(f"Точність на тестовому наборі: {test_accuracy}")
```
