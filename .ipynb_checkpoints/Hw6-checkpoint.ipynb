```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, fetch_openml
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# os.environ["LOKY_MAX_CPU_COUNT"] = "1"  
```

Завантаження двовимірного датасету

```python
data_2d = pd.read_csv("data_2d.csv")
X_2d = data_2d.iloc[:, 1:].values
print(data_2d)
```

Завантаження MNIST датасету

```python
mnist = pd.read_csv("mnist.csv")
X_mnist = mnist.values    
X_mnist = mnist.iloc[:, 1:].values
print(mnist)
```

Застосування K-means для кластеризації

```python
def kmeans_clustering(X, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(X)
    return kmeans.labels_
```

Визначення оптимальної кількості кластерів

```python
def find_optimal_clusters(X, max_clusters=10):
    inertias = []
    for i in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=i, random_state=0)
        kmeans.fit(X)
        inertias.append(kmeans.inertia_)
    plt.plot(range(1, max_clusters + 1), inertias, marker='o')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
    plt.title('Elbow Method')
    plt.show()
```

Візуалізація результатів кластеризації

```python
def visualize_clusters_2d(X, labels, centers):
    # Визначаємо мінімуми та максимуми для встановлення меж контурного графіку
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
    
    # Побудова контурного графіку
    Z = KMeans(n_clusters=len(centers)).fit(np.c_[xx.ravel(), yy.ravel()]).labels_
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
    plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('K-means Clustering (2D)')
    plt.show()
```

Візуалізація результатів кластеризації для MNIST датасету

```python
def visualize_clusters_mnist(X, labels, pca_transformed):
    plt.scatter(pca_transformed[:, 0], pca_transformed[:, 1], c=labels, s=10, cmap='viridis')
    plt.colorbar(label='digit label', ticks=range(10))
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title('K-means Clustering (MNIST)')
    plt.show()
```

Оптимальна кількість кластерів

```python
find_optimal_clusters(X_2d)
```

Результат для двовимірного датасету

```python
n_clusters_2d = 2
labels_2d = kmeans_clustering(X_2d, n_clusters_2d)
centers_2d = KMeans(n_clusters=n_clusters_2d).fit(X_2d).cluster_centers_
visualize_clusters_2d(X_2d, labels_2d, centers_2d)
```

Стандартизуємо MNIST датасет та виконуємо PCA

```python
scaler = StandardScaler()
X_mnist_scaled = scaler.fit_transform(X_mnist)
pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(X_mnist_scaled)
```

Знаходимо оптимальну кількість кластерів для MNIST датасет

```python
find_optimal_clusters(X_mnist_scaled)
```

Результат для MNIST датасету

```python
n_clusters_mnist = 9 
labels_mnist = kmeans_clustering(X_mnist_scaled, n_clusters_mnist)
visualize_clusters_mnist(X_mnist_scaled, labels_mnist, pca_transformed)
```
